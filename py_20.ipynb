# MRO(Model Resolution Object):
- mro is used to find the order of execution in  multiple inheritance.

```python
class A:
    def show(self):
        print("a")
class B(A):
    def show(self):
        print("b")
class C(A):
    def show(self):
        print("c")
class D(B,C):
        pass
d = D()
d.show()
D.mro()
```

```python
class add:
    def __init__(self,a,b):
        self.a = a
        self.b = b
    def addition(self):
        print(self.a+self.b)
class sub(add):
    def subtraction(self):
        print(self.a-self.b)
class mult(sub):
    def multiplication(self):
        print(self.a*self.b)
class div(mult):
    def division(self):
        print(self.a/self.b)
x = div(3,4)
x.addition()
x.subtraction()
x.multiplication()
x.division()
div.mro()
```

# Encapsulation
- it means binding data(variable) and methods(functions) that operates the data into a single unit.

```python
class person:
    def __init__(self,name,age,salary):
        self.name = name
        self._age = age
        self.__salary = salary
    
    def set_age(self,age):
        self._age = age
    def get_age(self):
        print(self._age)

    def set_salary(self,salary):
        self.__salary = salary
    def get_salary(self):
        print(self.__salary)
s = person("sai",21,"10l")
s.get_age()
s.get_salary()
```

# Polymorphism
- polymorphism means one object or method is behaving differently according to its object calling.

```python
class cat:
    def sound(self):
        return "meow"
class dog:
    def sound(self):
        return "bark"
def animal_sound(animal):
    print(animal.sound())
d = dog()
c = cat()
animal_sound(d)
animal_sound(c)
```

# Abstraction
- Abstraction is the process of hiding the essential process from the user.

```python
from abc import ABC ,abstractmethod
class shape(ABC):
    @abstractmethod
    def area(self):
        pass
class rectangle(shape):
    def __init__(self,l,w):
        self.l = l
        self.w = w
    def area(self):
        print(self.l * self.w)
class circle(shape):
    def __init__(self,r):
        self.r = r 
    def area(self):
        print(f"{3.14 * self.r**2}")
r = rectangle(3,4)
c = circle(5)
c.area()
r.area()
```
